```python
columns = concrete_data.columns

fig, ax = plt.subplots(4, 2, figsize=(25, 50))
for i in range(len(columns) - 1) :
  ax[i // 2, i % 2].scatter(concrete_data[columns[i]], concrete_data[columns[-1]])
  ax[i // 2, i % 2].set_xlabel(columns[i])
  ax[i // 2, i % 2].set_ylabel(columns[-1])
plt.show()
```

```python
import sklearn.model_selection as sklt

concrete_data = concrete_data.to_numpy()

x = concrete_data[:, :-1]
y = concrete_data[:, -1]
x_train, x_test, y_train, y_test = sklt.train_test_split(
    x, y,
    random_state = 100,
    test_size = 0.01
)
```

```python
def predict(x, w, b) :
    return np.dot(x, w) + b
```

```python
def gradient_descent(x, y, learning_rate, iterations) :
    cost = []

    m, n = x.shape
    # n x 1
    w = np.zeros(n)
    b = 0.

    # dj_dw = 1 / m * sig {pred - true} * x_i
    for i in range(iterations) :

        dj_dw = (1 / m) * np.dot(
            x.T,
            predict(x, w, b) - y
        )
        dj_db = (1 / m) * np.sum(
            predict(x, w, b) - y
        )

        w = w - learning_rate * dj_dw
        b = b - learning_rate * dj_db

    return w, b
```

```python
w, b = gradient_descent(x_train, y_train, 0.000001, 10000)

y_pred = predict(x_test, w, b)

fig, ax = plt.subplots()
ax.plot(y_pred, 'o', label = "predicted")
ax.plot(y_test, 'd', label = "actual")
plt.legend()
plt.show()
```
