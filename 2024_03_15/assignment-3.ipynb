```python
import numpy as np
import pandas as pd
import sklearn.metrics as sklm
import sklearn.model_selection as sklms
import matplotlib.pyplot as plt

import sklearn.ensemble as skle
```

```python
hd_df = pd.read_csv("Heart_Disease .csv")
hd_df
```

```python
hd_df.info()
```

```python
column = hd_df.columns

for i in range(len(column)) :
    if (len(hd_df[column[i]].unique()) < 10) :
        print(f"{i}. {column[i]} column unique values : {hd_df[column[i]].unique()}")
```

```python
hd_df.replace(
    to_replace={'?' : np.nan},
    inplace = True
)

hd_df.dropna(axis=0, inplace=True)
```

```python
hd_df_filtered = pd.get_dummies(
    hd_df,
    columns= [
        column[2],
        column[6],
        column[10],
        column[11],
        column[12]
    ],
    dtype=int
)

hd_df_filtered.info()
```

```python
y = hd_df_filtered.iloc[:, 8]
x = hd_df_filtered.drop(axis= 1, columns="hd")
column = x.columns

x = x.to_numpy()
y = y.to_numpy()

print(f"x dim : {x.shape}")
print(f"y dim : {y.shape}")
```

```python
import sklearn.tree as sklt
```

```python
x_train, x_test, y_train, y_test = sklms.train_test_split(
    x, y,
    random_state= 100,
    test_size= 0.1
)

print("Train and Test data shapes")
print(f"x_train : {x_train.shape} x_test : {x_test.shape}")
print(f"y_train : {y_train.shape} y_test : {y_test.shape}")

clf_dt = sklt.DecisionTreeClassifier(
    criterion="gini"
)

clf_dt.fit(x_train, y_train)
```

```python
y_pred = clf_dt.predict(x_test)

print(sklm.classification_report(y_pred, y_test))

cm_dt = sklm.confusion_matrix(y_test, y_pred)

plt.imshow(cm_dt, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Decision Tree Confusion Matrix')
plt.colorbar()
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
dt_plot = sklt.plot_tree(clf_dt, max_depth=2, fontsize=10, feature_names=list(column))
plt.show()
```

```python
crs_val_scr = sklms.cross_val_score(clf_dt, x, y, cv=5)

plt.plot(range(1, 6), crs_val_scr, marker='o')
plt.xlabel('Fold')
plt.ylabel('Cross-Validation Score')
plt.title('Decision Tree Cross-Validation Score')
plt.show()
```

```python
clf_rf = skle.RandomForestClassifier()
clf_rf.fit(x, y)
```

```python
y_rf_pred = clf_rf.predict(x_test)

print(sklm.classification_report(y_rf_pred, y_test))
```

```python
feature_importance = clf_rf.feature_importances_
print("Feature Importance:")

for feature, importance in zip(column, feature_importance):
    print(f"{feature}  : {importance}")
```

```python
cf_mt = sklm.confusion_matrix(y_true=y_test, y_pred=y_rf_pred)

plt.imshow(cm_dt, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Decision Tree Confusion Matrix')
plt.colorbar()
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
crs_val_scr = sklms.cross_val_score(clf_rf, x, y, cv=5)

plt.plot(crs_val_scr, marker='o')
plt.xlabel('Fold')
plt.ylabel('Cross-Validation Score')
plt.title('Random Forest Cross-Validation Score')
plt.show()
```

```python
n_trees_range = [10, 20, 50, 75, 100, 120, 150, 175]
for n_trees in n_trees_range:
    rf_model_n = skle.RandomForestClassifier(n_estimators=n_trees, random_state=42)
    cv_score_n = sklms.cross_val_score(rf_model_n, x, y, cv=5).mean()
    print(f"Number of Trees: {n_trees}, Mean CV Score: {cv_score_n}")
```

```python
max_depth_range = [None, 1, 2, 3, 4, 5, 10, 20, 30]
for max_depth in max_depth_range:
    rf_model_depth = skle.RandomForestClassifier(max_depth=max_depth, random_state=42)
    cv_score_depth = sklms.cross_val_score(rf_model_depth, x, y, cv=5).mean()
    print(f"Max Depth: {max_depth}, Mean CV Score: {cv_score_depth}")
```

```python
n_features_range = [2, 3, 5, 8, 10, 'sqrt', 'log2']
for n_features in n_features_range:
    rf_model_features = skle.RandomForestClassifier(max_features=n_features, random_state=42)
    cv_score_features = sklms.cross_val_score(rf_model_features, x, y, cv=5).mean()
    print(f"Number of Features: {n_features}, Mean CV Score: {cv_score_features}")
```
