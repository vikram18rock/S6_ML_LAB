```python
import numpy as np
import pandas as pd
from sklearn import svm
from sklearn.model_selection import train_test_split
import sklearn.metrics as sklm
import matplotlib.pyplot as plt
from sklearn.inspection import DecisionBoundaryDisplay
import sklearn.inspection
```

```python
iris_df = pd.read_csv("Iris.csv")
iris_df
```

```python
iris_df.info()
```

> Here Species column is of string type

> Also the Id column is not needed

```python
iris_df.drop(axis=1, columns="Id", inplace=True)
```

```python
labels = iris_df.iloc[:, -1].unique()
iris_df.replace(
    labels, 
    [i for i in range(len(labels))], 
    inplace=True
)

iris_df.info()
```

```python
columns = iris_df.columns
for column in columns[:-1] :
    fig, ax = plt.subplots()
    ax.scatter(iris_df[column], iris_df.iloc[:, -1])
    ax.set_xlabel(column)
    ax.set_ylabel(columns[-1])
    plt.show()
```

```python
from sklearn.inspection import DecisionBoundaryDisplay
```

```python
for i in range(len(columns) - 2) :
    for j in range(i + 1, len(columns) - 1) :
        x, y = iris_df[[columns[i], columns[j]]], iris_df.iloc[:, -1]
        x_train, x_test, y_train, y_test  = train_test_split(x, y, random_state=200, test_size=0.1)

        kernels = ['linear', 'sigmoid']

        C= 1.0
        models = [
            svm.SVC(kernel="linear", C=C),
            svm.SVC(kernel="sigmoid", C=C)
        ]
        # training
        models[0].fit(x_train, y_train)
        models[1].fit(x_train, y_train)

        # predicting 
        y_pred_lin = models[0].predict(x_test)
        y_pred_sig = models[1].predict(x_test)

        disp = DecisionBoundaryDisplay.from_estimator(models[0],response_method="predict", X=x,
            xlabel=columns[i], ylabel=columns[j],
            alpha=0.5)

        disp.ax_.scatter(x.iloc[:, 0], x.iloc[:, 1], c=y, edgecolors='k')
        disp.ax_.set_title("Linear kernel")
        plt.show()
        print("Evaluation Metrics")
        print(sklm.classification_report(y_test, y_pred_lin, zero_division=1))

        disp = DecisionBoundaryDisplay.from_estimator(models[1],response_method="predict", X=x,
            xlabel=columns[0], ylabel=columns[1],
            alpha=0.5)
        
        disp.ax_.scatter(x.iloc[:, 0], x.iloc[:, 1], c=y, edgecolors='k')
        disp.ax_.set_title("sigmoid kernel")
        plt.show()
        print("Evaluation Metrics")
        print(sklm.classification_report(y_test, y_pred_sig, zero_division=1))
```
