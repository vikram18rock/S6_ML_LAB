> Also the Id column is not needed

```python
iris_df.drop(axis=1, columns="Id", inplace=True)
```

```python
labels = iris_df.iloc[:, -1].unique()
iris_df.replace(
    labels, 
    [i for i in range(len(labels))], 
    inplace=True
)

iris_df.info()
```

```python
from sklearn.inspection import DecisionBoundaryDisplay
```

```python
x, y = iris_df.iloc[:, : -1], iris_df.iloc[:, -1]
x_train, x_test, y_train, y_test  = train_test_split(x, y, random_state=200, test_size=0.3)

kernels = ['linear', 'sigmoid']

C= 1.0
models = [
    svm.SVC(kernel="linear", C=C),
    svm.SVC(kernel="sigmoid", C=C)
]
# training
models[0] = models[0].fit(x_train, y_train)

y_train_sig = y_train
y_train_sig[y_train_sig != 0] = 1

y_test_sig = y_test
y_test_sig[y_test_sig != 0] = 1

models[1] = models[1].fit(x_train, y_train_sig)

# predicting 
y_pred_lin = models[0].predict(x_test)
y_pred_sig = models[1].predict(x_test)

print(y_pred_lin)
print(y_pred_sig, y_test_sig)
```
