# Series

```python
import pandas as pd
import numpy as np

data = np.array(['g', 'e', 'e', 'k', 's'])

ser = pd.Series(data)
print(ser)
```

# Data Frames

```python
import pandas as pd
data = {'name' : ['Tom', 'nick', 'krish', 'jack'], 
        'age' : [20, 21, 22, 23]}

df = pd.DataFrame(data)
print(df)
```

# Column Selection

```python
import pandas as pd

data = {'name' : ['Tom', 'nick', 'krish', 'jack'],
        'age' : [20, 21, 22, 23],
        'address' : ['a', 'b', 'c', 'd']}

df = pd.DataFrame(data)

# selecting name, address columns
print(df[['name', 'address']])
```

# Row Selection

## loc method

```python
import pandas as pd

data = pd.read_csv("nba.csv", index_col= "Name")

first = data.loc["Avery Bradley"]

second = data.loc["R.J. Hunter"]

print
print(first, "\n\n")
print(second)
```

## iloc method

```python
row2 = data.iloc[3]
print(row2)

display(data.iloc[[0,2, 4, 7]])

display(data.iloc[1: 5])

data.loc[((data.Number == 0, ['Team', 'College']))]
```
